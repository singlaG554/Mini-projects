```python
import cv2
import os

def resize_video(input_path, output_path, scale_percent=50):
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print("❌ Cannot open video file.")
        return

    # Original dimensions
    original_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    original_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Calculate new dimensions
    new_width = int(original_width * scale_percent / 100)
    new_height = int(original_height * scale_percent / 100)

    # Define codec and VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, fps, (new_width, new_height))

    print("🔄 Resizing video...")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Resize frame
        resized_frame = cv2.resize(frame, (new_width, new_height), interpolation=cv2.INTER_AREA)
        out.write(resized_frame)

    cap.release()
    out.release()
    print(f"✅ Resized video saved to: {output_path}")

# Example usage
resize_video("/content/WhatsApp Video 2025-06-22 at 2.34.51 PM.mp4", "resized_video.avi")
```
