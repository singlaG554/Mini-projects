```python
#upload your images here
from google.colab import files
import os

# Create input_images folder if not exists
os.makedirs("input_images", exist_ok=True)

# Upload files
uploaded = files.upload()

# Move them into input_images folder
for fname in uploaded.keys():
    os.rename(fname, os.path.join("input_images", fname))
```

```python
import cv2
import os

def resize_images(input_folder, output_folder, scale_percent=50, max_images=100):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:max_images]

    for idx, image_name in enumerate(image_files):
        input_path = os.path.join(input_folder, image_name)
        output_path = os.path.join(output_folder, image_name)

        # Read the image
        img = cv2.imread(input_path)
        if img is None:
            print(f"❌ Could not read {image_name}. Skipping.")
            continue

        # Calculate new dimensions
        width = int(img.shape[1] * scale_percent / 100)
        height = int(img.shape[0] * scale_percent / 100)
        resized_img = cv2.resize(img, (width, height), interpolation=cv2.INTER_AREA)

        # Save resized image
        cv2.imwrite(output_path, resized_img)
        print(f"[{idx+1}] Resized: {image_name}")

    print("✅ Resizing completed for", len(image_files), "images.")

# Example usage
resize_images("input_images", "resized_images")
```
