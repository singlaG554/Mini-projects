```python
import cv2
import os

def convert_images_to_grayscale(input_folder, output_folder, max_images=100):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:max_images]

    for idx, image_name in enumerate(image_files):
        input_path = os.path.join(input_folder, image_name)
        output_path = os.path.join(output_folder, image_name)

        # Read the image
        img = cv2.imread(input_path)

        # Convert to grayscale
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Save the grayscale image
        cv2.imwrite(output_path, gray_img)
        print(f"[{idx+1}] Converted: {image_name}")

    print("✅ Conversion completed for", len(image_files), "images.")

# Example usage
convert_images_to_grayscale("input_images", "output_images")
```
